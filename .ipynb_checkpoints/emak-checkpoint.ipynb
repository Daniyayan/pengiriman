```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# memasukan dataset 
df = pd.read_csv('dataset_ecommerce (8).csv')
```

1. Lakukan EDA

```python
print(df.info())
print(df.describe())
```

```python
# Menghitung frekuensi nilai unik
print(df['courier_delivery'].value_counts())
print(df['city'].value_counts())
print(df['type_of_delivery'].value_counts())
```

```python
# Distribusi status pengiriman
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='type_of_delivery')
plt.title('Distribution of Delivery Status')
plt.show()
```

```python
# Distribusi ke kota mana 
plt.figure(figsize=(10, 6),34)
sns.countplot(data=df, x='courier_delivery')
plt.title('Distribution of Couriers')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualisasi Data
sns.countplot(x='ontime', data=df)
plt.title('Distribusi Ontime')
plt.show()
```

2. Kurir mana yang paling sering ontime dan paling sering delay ?

```python
# mengechek berapa kali muncul untuk pengiriman yang tepat waktu dan tidak tepat waktu dalam DataFrame 
ontime_stats = df.groupby('courier_delivery')['ontime'].value_counts(normalize=True).unstack()
print(ontime_stats)
```

3. Cek, setiap kurir pernah berapa kali delay dan berapa kali ontime?

```python
courier_ontime_counts = df.groupby('courier_delivery')['ontime'].value_counts()
print(courier_ontime_counts)
```

4. "Type of delivery" mana dari setiap kurir yang sering delay dan sering ontime?

```python
type_delivery_stats = df.groupby(['courier_delivery', 'type_of_delivery'])['ontime'].value_counts(normalize=True).unstack()
print(type_delivery_stats)
```

5. Di Kota mana yang sering mengalami delay?

```python
city_delay_counts = df[df['ontime'] =34= 'Delayed']['city'].value_counts()
print(city_delay_counts)
```

6. Di kota mana yang sering merasakan ontime?

```python
city_ontime_counts = df[df['ontime'] == 'Ontime']['city'].value_counts()
print(city_ontime_counts)
```

7. Secara keseluruhan, produk rating berapa yang paling banyak dikirim?

```python
most_common_rating = df['product_rating'].mode().values[0]
print(f"Rating produk paling banyak dikirim: {most_common_rating}")
```

8. Di setiap kota, kurir mana yang sering delay?

```python
city_courier_delay_stats = df[df['ontime'] =34= 'Delayed'].groupby('city')['courier_delivery'].value_counts()
print(city_courier_delay_stats)
```

9. Di setiap kota, kurir mana yang sering ontime?

```python
city_courier_ontime_stats = df[df['ontime'] =34= 'Ontime'].groupby('city')['courier_delivery'].value_counts()
print(city_courier_ontime_stats)
```
